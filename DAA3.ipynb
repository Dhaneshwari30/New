```python
def fractional_knapsack():
    # Item weights and their corresponding values
    weights = [10, 20, 30]
    values = [60, 100, 120]
    capacity = 50  # Maximum capacity of the knapsack
    res = 0.0      # Total value (use float for fractional values)

    # Pair each item as (weight, value) and sort by value/weight ratio in descending order
    for weight, value in sorted(zip(weights, values), key=lambda x: x[1] / x[0], reverse=True):
        if capacity <= 0:  # If knapsack is full, stop
            break

        if weight > capacity:  # If item cannot fit completely, take fraction
            res += capacity * (value / weight)
            capacity = 0
        else:  # If item fits completely, take the whole item
            res += value
            capacity -= weight

    print("Maximum value in Knapsack =", res)


if __name__ == "__main__":
    fractional_knapsack()
```
