```python
# Function to generate Fibonacci series using Recursion
def recursive_fibonacci(n):
    if n <= 1:
        return n
    else:
        return recursive_fibonacci(n - 1) + recursive_fibonacci(n - 2)

# Function to generate Fibonacci series using Iteration (Non-Recursive)
def non_recursive_fibonacci(n):
    first = 0
    second = 1

    print(first)
    print(second)

    for i in range(2, n):
        third = first + second
        print(third)
        first = second
        second = third

# Main Program
if __name__ == "__main__":
    n = 10  # Number of terms to display

    print("Recursive Fibonacci Series:")
    for i in range(n):
        print(recursive_fibonacci(i))

    print("\nNon-Recursive Fibonacci Series:")
    non_recursive_fibonacci(n)
```
